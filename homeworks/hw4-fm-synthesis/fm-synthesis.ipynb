```python
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt
from scipy.io import wavfile

import warnings
warnings.filterwarnings("ignore")
```

## Phase Modulation
$y(t) = sin(\omega_c t + \Delta\theta sin(\omega_m t))$

```python
t = np.linspace(0, 1, 2048)
fc = 10
fm = 1
I = 100
modulation_signal = np.sin(2 * np.pi * fm * t)
pfm_signal = np.sin(2 * np.pi * fc * t + I * modulation_signal)
plt.plot(t, pfm_signal)
plt.show()
```

```python
plt.plot(plt.specgram(pfm_signal))
```

## Exponential FM 
$y(t) = sin(\omega_c2^{Vsin(\omega_mt)}t)$


```python
t = np.linspace(0, 1, 2048)
fc = 100
fm = 1
V = 20
modulation_signal = np.sin(2 * np.pi * fm * t)
efm_signal = np.sin(2 * np.pi * fc * 2 ** (V * modulation_signal) * t)
plt.plot(t, efm_signal)
plt.show()
```

```python
plt.plot(plt.specgram(efm_signal))
```

```python
import numpy as np
import matplotlib
import scipy.signal as signal
import matplotlib.pyplot as plt
from scipy.io import wavfile
from IPython.display import Audio

matplotlib.rcParams['figure.figsize'] = (15, 7)
```

```python
# load data
fs, data = wavfile.read("lick.wav")
fs_target = 8000

def lowpass_fir(sample_rate, x):
    nyq_freq = sample_rate * 0.5
    transition_width = 1e2 / nyq_freq
    
    stop_attenuate_db = 60.0
    N, beta = signal.kaiserord(stop_attenuate_db, transition_width)
    cutoff_freq = 1e3
    taps = signal.firwin(N, cutoff_freq/nyq_freq, window=("kaiser", beta))
    return signal.lfilter(taps, 1.0, x)

# convert to mono
input_signal = data.sum(axis=1) / 2

# filter
filtered_signal = lowpass_fir(fs, input_signal)

# decimation
downsampled_signal = filtered_signal[::fs//fs_target]

print(f"Original signal sample count: {input_signal.size}\nDownsampled signal sample count: {downsampled_signal.size}")

Audio(downsampled_signal, rate=fs_target)
```

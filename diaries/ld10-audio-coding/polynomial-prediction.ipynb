```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rcParams['figure.figsize'] = (15, 8)
```

```python
n_samples = 30
t = np.linspace(0, 1, n_samples)

sinusoid = np.sin(2 * np.pi * t)
plt.scatter(t, sinusoid)
```

```python
plt.figure()

legend_list = ["Original sinusoid"]
plt.scatter(t, sinusoid)
plt.axvline(t[-1], color="r", linestyle="--", label='_nolegend_')
plt.axhline(sinusoid[-1], color="r", linestyle="--", label='_nolegend_')
for deg in range(1, 6):
    coeffs = np.polyfit(t[:-1], sinusoid[:-1], deg)
    prediction = np.polyval(coeffs, t)
    print("="*100)
    print(f"Degree {deg}\nAbsolute prediction error: {np.abs(sinusoid[-1] - prediction[-1]):.3f}")
    plt.plot(t, prediction)
    legend_list.append(f"Degree {deg}")

plt.legend(legend_list)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal

duration = 2
fs = 200
samples = fs * duration
t = np.arange(samples) / fs
#generating signal with modulation
original_signal = np.sin(100*np.pi*t) * np.cos(2*np.pi*t)

# performing envelope estimation
analytic = signal.hilbert(original_signal)
envelope = np.abs(analytic)

plt.figure(figsize=(10, 5))
plt.grid()
plt.plot(original_signal)
plt.plot(envelope)
plt.xlabel("Sample Number")
plt.ylabel("Amplitude")
plt.legend(["Signal", "Envelope"])
plt.savefig("envelope.png")
plt.show()
```
